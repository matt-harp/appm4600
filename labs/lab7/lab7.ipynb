```python
from matplotlib import pyplot as plt
import numpy as np

from labs.lab7.interp import eval_lagrange, evalDDpoly, dividedDiffTable
from labs.lab7.monomial_interp import Vandermonde, eval_monomial

a = -1
b = 1
N = 17
f = lambda x: 1/(1+(10*x)**2)

# interpolating nodes
xint = []
for j in range(1,N+1):
    xint.append(np.cos((((2*j)-1)*np.pi)/(2*N)))
xint.append(0)
xint = np.array(xint)
yint = f(xint)

V = Vandermonde(xint, N)
Vinv = np.linalg.inv(V)
coef = Vinv @ yint

# evaluation points
Neval = 1000
xeval = np.linspace(a, b, Neval+1)
yeval = f(xeval)

plt.plot(yeval, label='f(x)')
y_mon = eval_monomial(xeval, coef, N, Neval)
plt.plot(y_mon, label='monomial')

y_lag = np.zeros(Neval+1)
for i in range(Neval+1):
    y_lag[i] = eval_lagrange(xeval[i], xint, yint, N)
plt.plot(y_lag, label='lagrange')

y = np.zeros((N+1,N+1))
y[:,0] = yint
y_ndd = np.zeros(Neval + 1)
for i in range(Neval + 1):
    y_ndd[i] = evalDDpoly(xeval[i], xint, dividedDiffTable(xint, y, N+1), N)
plt.plot(y_ndd, label='ddpoly')

plt.legend()
plt.show()
```
