```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

delta = np.logspace(-16, 0, 17)

x1 = np.pi
x2 = 10**6

def true(x):
    return np.cos(x + delta) - np.cos(x)

def other(x):
    return -2 * np.sin(delta / 2)*np.sin(x + delta/2)

plt.semilogx(delta, (other(x1) - true(x1)))
plt.show()
plt.semilogx(delta, (other(x2) - true(x2)))
plt.show()
```
