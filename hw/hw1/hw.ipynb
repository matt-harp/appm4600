```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

delta = np.logspace(-16, 0, 17)

x1 = np.pi
x2 = 10**6

def taylor(x):
    return -delta * np.sin(x) - delta**2/2

def true(x):
    return -2 * np.sin(delta / 2)*np.sin(x + delta/2)

plt.semilogx(delta, (true(x1) - taylor(x1)))
plt.show()
plt.semilogx(delta, (true(x2) - taylor(x2)))
plt.show()
```
