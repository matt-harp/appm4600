```python
import numpy as np
import matplotlib.pyplot as plt

maclaurin = lambda x: x - x**3/6 + x**5/120
p33 = lambda x: (x - (7/60)*x**3)/(1 + (1/20)*x**2)
p24 = lambda x: (x)/(1-(1/6)*x**2 + (7/360)*x**4)
p42 = lambda x: (x - (7/60)*x**3)/(1 + (1/20)*x**2)

x = np.linspace(0, 5, 1000)
yreal = np.sin(x)

merr = np.abs(maclaurin(x) - yreal)
_33err = np.abs(p33(x) - yreal)
_24err = np.abs(p24(x) - yreal)
_42err = np.abs(p42(x) - yreal)

plt.semilogy(x, merr, label='Maclaurin')
plt.semilogy(x, _33err, label='3-3')
plt.semilogy(x, _24err, label='2-4')
plt.semilogy(x, _42err, label='4-2')
plt.legend()
plt.xlabel('x')
plt.ylabel('logy')
plt.show()
```
