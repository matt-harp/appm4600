```python
import numpy as np
import matplotlib.pyplot as plt

def curve(R, dr, f, p):
    theta = np.linspace(0, 2*np.pi, 1000)
    x = R*(1+dr*np.sin(f*theta+p))*np.cos(theta)
    y = R*(1+dr*np.sin(f*theta+p))*np.sin(theta)
    plt.axis('equal')
    plt.plot(x, y)
```

```python
curve(1.2, 0.1, 15, 0)
```

```python
import random

for i in range(1, 11):
    curve(i, 0.05, 2 + i, random.uniform(0, 2))
```
