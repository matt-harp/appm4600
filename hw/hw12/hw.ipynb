```python
import numpy as np
from scipy.linalg import hilbert, invhilbert

def powerIt(A, Nmax):
    n = A.shape[0]
    q = np.random.rand(n)
    q = q/np.linalg.norm(q)
    lam = np.zeros(Nmax)
    for j in range(Nmax):
        z = A@q
        q = z/np.linalg.norm(z)
        lam[j] = q.T@A@q

        if j > 0 and np.abs(lam[j] - lam[j-1]) < 1e-5:
            lam = lam[:j+1]  # Trim unused part of lam
            break
    return lam, q

# for n in np.arange(4,24,4):
#     A = hilbert(n)
#     lam,q = powerIt(A, 1000)
#     print(f'n={n}: iter= lam={lam[-1]}')

A = hilbert(16)
lam, q = powerIt(A, 100)
print(lam[-1])

np.linalg.eig(A)
```
